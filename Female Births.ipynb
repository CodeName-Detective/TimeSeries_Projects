```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("daily_Female_Births_Dataset.csv")
df
```

```python
df.Date.dtype
```

### Convert Date field from object to Datetime

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.Date.dtype
```

```python
df.Date
```

### Set Date Column as Index

```python
X = df.set_index('Date')

X
```

```python
split = int(len(X)/2)
X1,X2 = X[:split],X[split:]
mean1,mean2 = X1.mean(),X2.mean()
var1,var2 = X1.var(),X2.var()
print('mean1=%f, mean2=%f' %(mean1,mean2))
print('var1=%f, var2=%f' %(var1,var2))
```

### Graphs

```python
plt.figure(figsize=(20,5))
plt.plot(df['Date'],df['Births'])
```

# Hypothesis Testing - Augmented Dickey Fuller Test

- Null Hypothesis = The time series is non-stationery

```python
from statsmodels.tsa.stattools import adfuller
test_result = adfuller(df['Births'])
```

```python
test_result
```

```python
print('p value of the test is %f' % test_result[1])
```

- Reject Null Hypothesis

# Decompsoing Timeseries

```python
from statsmodels.tsa.seasonal import seasonal_decompose
X1 = df.set_index('Date')
decompse_result=seasonal_decompose(X1['Births'],model='multiplicative')
decompse_result.plot()
plt.figure(figsize=(20,5))
plt.show()
```

# ACF(Auto Correlation Function) and PACF(Partial Auto Correlation Function)- Plot

```python
from statsmodels.tsa.stattools import acf,pacf
lag_acf = acf(df['Births'],nlags=20)
lag_pacf = pacf(df['Births'],nlags=20,method='ols')
```

```python
plt.figure(figsize=(20,10))
#Plot ACF:
plt.subplot(121)
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(df['Births'])),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(df['Births'])),linestyle='--',color='gray')
plt.title('Auto Correlation Function')


#Plot PACF:
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(df['Births'])),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(df['Births'])),linestyle='--',color='gray')
plt.title('partial Auto Correlation Function')

plt.tight_layout()
```

# ARIMA Model

```python
from statsmodels.tsa.arima_model import ARIMA
train,test = X[:360],X[350:]
train['Births'].dropna(inplace=True)
test['Births'].dropna(inplace=True)
arima_model = ARIMA(train['Births'],order=(2,0,2))
results_arima = arima_model.fit(disp=-1)
```

```python
len(results_arima.forecast(steps=14)[0]),len(test['Births'])
```

```python
result_data = {'predict':results_arima.forecast(steps=14)[0],'Actual':test['Births']}
predicted_df = pd.DataFrame(result_data)
predicted_df
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(predicted_df['Actual'],predicted_df['predict'])
mse
```
